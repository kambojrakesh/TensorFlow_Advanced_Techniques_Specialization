```python
import matplotlib.pylab as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier  
from sklearn.datasets import load_breast_cancer

#Load the Dataset
cancer_data = load_breast_cancer()

#Logistic Regression
X_train, X_test, y_train, y_test = train_test_split(cancer_data.data, cancer_data.target, stratify=cancer_data.target, random_state=42)

#KNN Classifier 
X_train, X_test, y_train, y_test = train_test_split(cancer_data.data, cancer_data.target, stratify=cancer_data.target, random_state=66)

training_accuracy = []
val_accuracy = []

#trying KNN for diffrent k nearest neighbor from 1 to 16
neighbors_range = range(1,16)

for no_neighbors in neighbors_range:
    knn = KNeighborsClassifier(n_neighbors=no_neighbors)
    knn.fit(X_train,y_train)
    training_accuracy.append(knn.score(X_train, y_train))
    val_accuracy.append(knn.score(X_test, y_test))
 
print("Accuracy of the training{:3f}".format(training_accuracy[5]))
print("Accuracy of the test {:3f}".format(val_accuracy[5]))

plt.plot(neighbors_range,training_accuracy, label='Accuracy of the training set')
plt.plot(neighbors_range ,val_accuracy, label='Accuracy of the validation set')
plt.ylabel('Accuracy')
plt.xlabel('Number of Neighbor')
plt.legend()
```
